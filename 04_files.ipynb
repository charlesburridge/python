```python
#Import the numpy tools
%matplotlib inline
import numpy
import matplotlib.pyplot

filenames=glob.glob('data/inflammation*.csv')
#Import data


for fn in filenames[4:7]:
    data= numpy.loadtxt(fname=fn,delimiter=',')

    #Make blank figure
    fig=matplotlib.pyplot.figure(figsize=(10,3))
    axis1=fig.add_subplot(1,3,1)
    axis2=fig.add_subplot(1,3,2)
    axis3=fig.add_subplot(1,3,3)

    axis1.set_ylabel("average")
    axis1.plot(data.mean(axis=0))
    axis1.set_xlabel("days")

    axis2.set_ylabel('max')
    axis2.plot(data.max(axis=0))
    axis2.set_xlabel("days")

    axis3.set_ylabel("min")
    axis3.plot(data.min(axis=0))
    axis3.set_xlabel("days")

    matplotlib.pyplot.show()
```
